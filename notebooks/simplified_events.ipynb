```python
import pandas as pd
from tqdm import tqdm
```

# Import all match IDs

```python
matches_df = pd.read_csv('../data/matches.csv')
matches_df.head()
```

```python
match_ids = list(sorted(set(matches_df['match_id'])))
```

```python
selected_columns = [
    'carry_end_location',
    'dribble_outcome',
    'foul_won_penalty',
    'id',
    'index',
    'location',
    'match_id',
    'out',
    'pass_angle',
    'pass_assisted_shot_id',
    'pass_cross',
    'pass_end_location',
    'pass_goal_assist',
    'pass_height',
    'pass_length',
    'pass_outcome',
    'pass_recipient',
    'pass_recipient_id',
    'pass_shot_assist',
    'pass_type',
    'play_pattern',
    'player',
    'player_id',
    'position',
    'possession',
    'possession_team',
    'possession_team_id',
    'shot_end_location',
    'shot_freeze_frame',
    'shot_key_pass_id',
    'shot_outcome',
    'shot_statsbomb_xg',
    'shot_type',
    'tactics',
    'team',
    'team_id',
    'timestamp',
    'type'
]
```

```python
len(selected_columns)
```

```python
for i, id in enumerate(tqdm(match_ids)):
    df = pd.read_csv(f'../data/events/{id}.csv')
    available_columns = [c for c in selected_columns if c in df.columns]
    empty_columns = [c for c in selected_columns if c not in df.columns]
    df = df[available_columns]
    df = df.sort_values(by=['index'], ascending=True)
    for c in empty_columns:
        df[c] = ""
    df.to_csv(f'../data/events_simplified/{id}.csv')
```
