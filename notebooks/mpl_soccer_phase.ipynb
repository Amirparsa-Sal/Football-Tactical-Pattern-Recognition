```python
import pandas as pd
from ftpr.entities import Phase
from ftpr.visualization import PhaseVisualizer
import os
```

### Set Variables

```python
phases_dir = os.path.join('..', 'data', 'team_phases')
team = 'Manchester City'
phase_id = 105
```

### Load The Phase

```python
df = pd.read_csv(os.path.join(phases_dir, f'{team}.csv'))
print('Total phases:', len(df['phase_id'].unique()))
```

```python
df_phase = df[df['phase_id'] == phase_id]
df_phase.head()
```

```python
phase = Phase(df_phase)
phase.get_summary()
```

### Filter Static Events

```python
phase = phase.filter_static_events()
phase.get_summary()
```

### Visualize The Phase

```python
visualizer = PhaseVisualizer(figsize=(16, 11))
events_config = {
    'Pass': {
        'color': '#00ffff'
    },
    'Carry': {
        'color': '#00aa00',
        'dashed': True
    },
    'Shot': {
        'color': '#aa0000'
    } 
}

location_columns = ['location', 'pass_end_location', 'carry_end_location', 'shot_end_location']
phase = phase.split_locations(location_columns)

fig, ax = visualizer.plot_phase(phase, events_config)
```
