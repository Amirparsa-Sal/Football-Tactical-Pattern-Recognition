```python
import pandas as pd
from mplsoccer import Pitch
from ftpr.entities import Phase
```

```python
team_name = 'Manchester City'
df = pd.read_csv(f'../data/team_phases/{team_name}.csv')
print('Total phases:', len(df['phase_id'].unique()))
df.head()
```

```python
phase_id = 105
df_phase = df[df['phase_id'] == phase_id]
df_phase.head()
```

```python
phase = Phase(df_phase)
phase.get_summary()
```

```python
phase = phase.filter_static_events()
phase.get_summary()
```

```python
def plot_path(pitch, ax, df, event_type, color='#000000', dashed=False):
    if dashed:
        return pitch.lines(df.location_x, df.location_y, df[f'{event_type.lower()}_end_location_x'], 
                           df[f'{event_type.lower()}_end_location_y'], lw=5, transparent=True, 
                           label=f'{event_type.capitalize()}', color=color, ax=ax, linestyle='--')
        
    return pitch.arrows(df.location_x, df.location_y, df[f'{event_type.lower()}_end_location_x'], 
                        df[f'{event_type.lower()}_end_location_y'], label=f'{event_type.capitalize()} Path',
                        color=color, ax=ax, width=3)
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

location = None

event_types = phase['type'].unique()
location_columns = ['location', 'pass_end_location', 'carry_end_location', 'shot_end_location']

# Plot event locations
phase = phase.split_locations(location_columns)
for et in event_types:
    df_event = phase[phase['type'] == et]
    sc3 = pitch.scatter(df_event.location_x, df_event.location_y, s=600, ax=ax, label=et)

# Plot passes path
df_passes = phase[df.type == 'Pass']
plot_path(pitch, ax, df_passes, 'pass', color='#00ffff')

# Plot carries path
df_carries = phase[df.type == 'Carry']
plot_path(pitch, ax, df_carries, 'carry', color='#00aa00', dashed=True)

# Plot carries path
df_shots = phase[df.type == 'shot']
plot_path(pitch, ax, df_shots, 'shot', '#aa0000')

# Plot the order of events
location_x, location_y =  phase.iloc[0]['location_x'], phase.iloc[0]['location_y']
pitch.annotate('1', (location_x, location_y), va='center', ha='center', color='white', fontsize=15, ax=ax)

for i in range(1, len(phase)):
    row = phase.iloc[i]
    if row['location_x'] != location_x or row['location_y'] != location_y:
        pitch.annotate(f'{i + 1}', (row.location_x, row.location_y), va='center', ha='center', color='white',
                fontsize=15, ax=ax)
    location_x, location_y = row['location_x'], row['location_y']
    
# Enable the legend
ax.legend(loc='upper left', labelspacing=1.5)
```
