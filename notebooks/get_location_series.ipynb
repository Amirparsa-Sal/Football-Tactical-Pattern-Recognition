```python
import numpy as np
import pandas as pd
from ftpr.entities import Phase
```

```python
team_name = 'Manchester City'
df = pd.read_csv(f'../data/team_phases/{team_name}.csv')
```

```python
phase_id = 105
phase_df = df[df['phase_id'] == phase_id]
phase_df[['location', 'pass_end_location', 'carry_end_location', 'type', 'timestamp']]
```

```python
phase = Phase(phase_df)
filtered_phase = phase.filter_static_events()
filtered_phase.get_summary()
```

```python
filtered_phase.get_location_series(location_columns=['location'])
```

```python
filtered_phase.get_location_series(location_columns=['location'], remove_duplicates=True)
```

```python
splitted_phase = filtered_phase.split_locations(location_columns=['location', 'pass_end_location', 'carry_end_location'])
```

```python
splitted_phase.get_summary(splitted_location=True)
```

```python
splitted_phase.get_location_series(location_columns=['location', 'pass_end_location'], splitted_locations=True)
```
